## 샘플 규칙

- 만약 탑승객의 pclass(좌석클래스)가 1등석이고 sex(성별)이 female(여성)이라면 생존 확률이 높을 것이다.

- 만약 탑승객의 age(나이)가 9.5살보다 어리고 sibsp(함께 탑승한 형제 또는 배우자의 수)가 2.5 명보다 많다면 생존 확률이 높을 것이다.

```python
titanic_df = pd.read_csv('titanic_train.csv')
df = titanic_df.loc[:, ['Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch','Ticket', 'Cabin', 'Embarked']]
print('titanic 변수 type:', type(df))
df
```

```python
#두 가지 조건을 만족하는 데이터 필터링

is_pclass = df['Pclass']== 1
is_sex = df['Sex'] == 'female'
survived_df = df[is_pclass & is_sex]
survived_df
```

```python
#Survived 0과 1의 개수 세기
survived_df['Survived'].isin(['1'])
#bool의 True, False 값 찾기 아아아아악!!!!!!!!!!!!!!!!!!!!!!!!!!!!
survived_df['Survived'].value_counts()
```

```python
#두 가지 조건을 만족하는 데이터 필터링

is_age = df['Age'] < 9.5
is_sibsp = df['SibSp'] > 2.5
ifsurvived_df = df[is_age & is_sibsp]
ifsurvived_df
```
